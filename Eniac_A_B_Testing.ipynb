```python
Click = [512, 281, 527, 193]
No_click = [24814, 24466, 24349, 25040]

observed_results = pd.DataFrame([Click, No_click],
                                  columns = ["A", "B", "C", "D"],
                                  index = ["Click", "No-click"])
observed_results
```

### Once we have our table. Let's calculate the rest with formulas ;)

```python
chisq, pvalue, df, expected = chi2_contingency(observed_results) #observed_result is our table
```

```python
chisq
```

```python
pvalue
```

```python
df
```

```python
expected
```

```python
alpha = 0.05
```

```python
if pvalue > alpha:
  print("The p-value is larger than alpha.")
else:
  print("The p-value is smaller than alpha.")
```

```python
if pvalue > alpha:
  print("Do not reject the null hypothesis")
else:
  print("Reject the null hypothesis")
```


Since the p-value is smaller than alpha, we will reject the Null Hypothesis. Therefore, the four versions have not performed equally well.

But which version is the best one?

Let's decide to look at the click-through rates of the different versions. They might help us to declare a winner.

```python
# calculate the click through rate for each version
ctr_a = observed_results.loc["Click", "A"] / sum(observed_results["A"])
ctr_b = observed_results.loc["Click", "B"] / sum(observed_results["B"])
ctr_c = observed_results.loc["Click", "C"] / sum(observed_results["C"])
ctr_d = observed_results.loc["Click", "D"] / sum(observed_results["D"])

# create a table with all four versions and their click-through rates.
ctrs = pd.DataFrame({"version": ["A", "B", "C", "D"],
                     "CTR": [ctr_a, ctr_b, ctr_c, ctr_d]})

ctrs.sort_values("CTR", ascending=False)
```

Version C exhibits a significantly higher click-through rate compared to versions B and D, whereas the difference between versions A and C is relatively minor.

Given the similarity between versions A and C, we will declare both versions as the victors with an enphasis on the C.
